```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
df = pd.read_csv('iris.csv')
```

```python
print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
mode_value = df['Species'].mode()
df.fillna({'Species': mode_value[0] if len(mode_value) > 0 else 'Missing'}, inplace=True)
```

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=df)
plt.show
```

```python
plt.hist(df['SepalLengthCm'])
plt.show()
```

```python
sns.boxplot(x='SepalLengthCm', data=df)
plt.show()
```
